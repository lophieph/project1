```python
import pandas as pd
import numpy as np
import datetime
```

```python
sales = pd.read_csv('sales.csv')
prices = pd.read_csv('prices.csv')
```

```python
sales.info()
prices.info()
```

```python
sales['ordered_at'] = sales['ordered_at'].apply(pd.to_datetime)
prices['updated_at'] = prices['updated_at'].apply(pd.to_datetime)
```

```python
sales.info()
prices.info()
```

```python
sales = sales.sort_values('ordered_at')
prices = prices.sort_values('updated_at')
```

```python
merge = pd.merge_asof(sales, prices, left_on = 'ordered_at', right_on = 'updated_at', by = 'product_id', direction = 'backward')
merge
```

```python
merge_table = pd.merge_asof(merge, prices, left_on = 'ordered_at', right_on = 'updated_at', by = 'product_id', direction = 'forward')
merge_table
```

```python

merge_table['price'] = merge_table['new_price_x'].fillna(merge_table['old_price_y'])
table = merge_table[['product_id', 'quantity_ordered', 'price']]
table
```

```python
table['revenue'] = table['quantity_ordered'] * table['price']
group_revenue = table.groupby(['product_id', 'price']).sum()
group_revenue
```

```python
total = group_revenue.groupby(['product_id']).sum()
total
```
